```python
import pandas as pd
import numpy as np
```

```python
file = "https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29"
GDP_table = pd.read_html(file)
GDP_table
```

```python
my_table = GDP_table[3]             # Select only the third table
my_table
```

```python
my_table.columns = range(my_table.shape[1])           # Rename column headers to numbers
```

```python
my_table
```

```python
table = my_table[[0,2]]                 # Select only Country and GDP
table
```

```python
table = table.iloc[1:11,:]         # Select top 10 countries and GDP
table
```

```python
table.columns  = ['Country' , 'GDP (Millions USD)']         # Rename header
table
```

```python
table['GDP (Millions USD)'] = table['GDP (Millions USD)'].astype(int)     #Use astype() to make values into integers
```

```python
table['GDP (Millions USD)'] = table['GDP (Millions USD)']/1000       #Convert the GDP value in Million USD to Billion USD
```

```python
table
```

```python
table['GDP (Millions USD)'] = np.round(table['GDP (Millions USD)'], 2)        #round to 2 decimal places
table
```

```python
table = table.rename(columns = {'GDP (Millions USD)' : 'GDP (Billions USD)'})       #Rename header
table
```

```python
table.to_csv('./Largest_economies.csv')
```
